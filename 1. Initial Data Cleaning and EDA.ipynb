```python
# Since the keys are duplicated in df_casualty and df_vehicle, we'll make a composite key to merge them first

df_vehicle['composite_key'] = df_vehicle.index + "_" + df_vehicle['vehicle_reference'].astype(str)
df_vehicle = df_vehicle.set_index('composite_key')

df_casualty['composite_key'] = df_casualty.index + "_" + df_casualty['vehicle_reference'].astype(str)
df_casualty = df_casualty.set_index('composite_key')
```

```python
df_merged = df_accident.join(df_vehicle_casualty, how='inner')
```

3. Decoding columns

```python
# Replace invalid codes with NaN or placeholders
columns_to_clean = {
    'age_of_casualty': [-1],
    'age_of_driver': [-1],
    'engine_capacity_cc': [-1],
    'generic_make_model': [-1],
    'speed_limit': [-1, 99]
}

for column, invalid_values in columns_to_clean.items():
    if column in df_merged.columns:
        df_merged[column] = df_merged[column].replace(invalid_values, [np.nan if val == -1 else "Unknown" for val in invalid_values])
```

```python
# Replace -1 with NaN in road number columns
road_number_columns = ['first_road_number', 'second_road_number']

for column in road_number_columns:
    if column in df_merged.columns:
        df_merged[column] = df_merged[column].replace(-1, np.nan)
```

```python
decoding_data = pd.read_excel(r"C:\Users\mnilo\OneDrive\Desktop\Courses\Master Project\dft-Decoding-Guide.xlsx",sheet_name="Sheet1")
```

```python
decoding_data
```

```python
decoding_dict = {}
for field in decoding_data['field_name'].unique():
    field_mapping = decoding_data[decoding_data['field_name'] == field][['code', 'description']]
    decoding_dict[field] = dict(zip(field_mapping['code'], field_mapping['description']))
```

```python
for column in decoding_dict.keys():
    if column in df_merged.columns:
        df_merged[column] = df_merged[column].map(decoding_dict[column]).fillna(df_merged[column])
```
